```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statistics
import os
```

```python
file_to_load = os.path.join("Resources", "Pyber_ride_data.csv")
pyber_ride_df = pd.read_csv(file_to_load)
pyber_ride_df
```

```python
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.show()
```

```python
# Set x-axis and tick locations.
x_axis = np.arange(len(pyber_ride_df))
tick_locations = [value for value in x_axis]
# Plot the data.
stdev = statistics.stdev(pyber_ride_df["Avg. Fare ($USD)"]) 
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", yerr=stdev, capsize=2)
plt.xticks(tick_locations, pyber_ride_df["Month"])
plt.show()
```

```python
pyber_ride_df.plot.bar(x="Month", y="Avg. Fare ($USD)", yerr=stdev, capsize=2)

```
